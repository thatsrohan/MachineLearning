```python
#Take the data and create dataframe
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/knn.csv')
df
```

```python
#Divide tha data into input and output
x=df.iloc[:,0:2].values
y=df.iloc[:,2]
```

```python
df.shape
```

```python
df.size
```

```python
k=5 #square root of total number of values or total number of rows
pred=[75,170]
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
```

```python
#model fitting
model.fit(x,y)
```

```python
y_pred=model.predict([pred])
y_pred
```
