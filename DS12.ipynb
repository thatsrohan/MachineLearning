```python
#Take the data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/knn.csv')
df
```

```python
#divide the data into input and output
x = df.iloc[:,0:2].values
y = df.iloc[:,2].values
```

```python
df.shape #25 rows and 3 cols
```

```python
df.size
```

```python
k = 5 #sqroot of total no of values/rows
pred = [75,170]
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = k,metric = 'euclidean')
```

```python
#model fitting
model.fit(x,y)
```

```python
#Predict the output
y_pred = model.predict([pred])
y_pred
```
