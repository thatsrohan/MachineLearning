```python
#INPUT
x=df.iloc[:,3:5].values
x
```

```python
#Visualisation
import matplotlib.pyplot as plt
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
#Here we have only one cluster  before applying clustering technique
```

```python
#Here our task is to find the number of cluster(k) into which our data will be divided
import numpy as np
np.sqrt(200) #200 is the total number of points/rows
#k value should not exceed the square root of the total number of points/rows
#K value should be in the range of 2 to 14
```

```python
#We need to find the number of clusters k
#1.ELBOW METHOD
#2.SILHOUTTE SCORE METHOD -very acccurate
```

```python
#ELBOW METHOD
from sklearn.cluster import KMeans
k=range(2,15)#2 is inclusive,15 is exclusive
sse =[] #empty list
#for i in range (2,15)
for i in k:
  model_demo=KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)#inertia calculates the sum of squared error
plt.scatter(k,sse)
plt.plot(k,sse)

```

```python
#Now that we have the elbow graph
#We wll now consider a point where elbow is more prominent
#We will consider k as 5 for now
```

```python
#2.Silhouette method
from sklearn.metrics import silhouette_score
k=range(2,15)
for i in k:
    model_demo=KMeans(n_clusters=i,random_state=0)
    model_demo.fit(x)
    y_pred=model_demo.predict(x)
    print(f'{i}Cluster,score={silhouette_score(x,y_pred)}')
    plt.bar(i,silhouette_score(x,y_pred))
    plt.show()
```

```python
#The numbers of clusters to be considered is 5
```

```python
#Applying a clusterer
model=KMeans(n_clusters=5,random_state=0)
model.fit(x)
```

```python
#Predicting the values
y=model.predict(x)
y #Here 4,3,1,2,0 are the cluster numbers
```
