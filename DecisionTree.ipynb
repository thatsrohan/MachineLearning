```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv")
print(df)


```

```python
inputs=df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
print(inputs)
```

```python
targets=df.Survived
print(targets)
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
inputs['Sex_n']=le_sex.fit_transform(inputs['Sex'])
inputs
```

```python
inputs_n=inputs.drop(['Sex'],axis='columns')
inputs_n
inputs_n=inputs_n.dropna
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs_n,targets)
```
