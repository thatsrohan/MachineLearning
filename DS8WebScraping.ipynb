```python
#WEB SCRAPING-EXTRActing DATA FROM THE WEB
#We will be scraping amazon custome reviews from amazon website
#Beautiful soup-tool used for web scraping
#url-https://www.amazon.in/Apple-iPhone-128GB-Space-Black/product-reviews/B0BDJ22G36/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.amazon.in/Apple-iPhone-128GB-Space-Black/product-reviews/B0BDJ22G36/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page=requests.get(url) #Requesting the contents of the above url
page
#Response[200] is correct,Response[500] is incorrect,try rerunning some more times
#Even after rerunning multiple times if you do not get the output,check for proper url
```

```python
page.content #This prints the HTML source code of the page
```

```python
soup=bs(page.content,'html.parser')
#html.parser makes our data into better readable format
soup
```

```python
#Never consider top positive and critical reviews
```

```python
 #Extracting the amazon customer review names
 names = soup.find_all('span',class_ = 'a-profile-name')[2:]
 names
```

```python
len(names)
```

```python
#Using the above list names create a dataframe
import pandas as pd
df=pd.DataFrame(names,columns=['Customer Name'])
df
```

```python
#Dropping the duplicates
df.drop_duplicates(ignore_index=True)
```

```python
#Scraping the review titles
r_title=soup.find_all('a',class_='review-title')
r_title
```

```python
len(r_title)
```

```python
#Data filtering
review_title=[] #empty list
for i in range(0,len(r_title)):
review_title.append(r_title[i].get_text()[1:-1])
review_title
```
