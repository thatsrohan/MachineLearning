```python
#4.Divide the data into input and output
x=df.iloc[:,2:4]
x
```

```python
y=df.iloc[:,4].values
y
```

```python
#5.Train and test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0) #random state is to avoid duplicates
#By derfault training variables get 75% of data and testing variables get 25% of data
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
#6.Normalisation/scaling-Done only for inputs
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#7.Apply a CLASSIFIER,regressor or clustere
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#8.Fit the model(we are giving out data to the logistic regression library)
model.fit(x_train,y_train)
```

```python
#9.Predict the output
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#10.Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
