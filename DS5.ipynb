```python
#MACHINE LEARNING-SUPERVISED LEARNING-REGRESSION-LINEAR REGRESSION
#UNIVARIATE/SIMPLE/SINGLE-1 Column as input and 1 column as output
#MULTIVARIATE/MULTIPLE-many columns as input and 1 column as output
#DATASET-https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
```

```python
#PREPROCESSING(OPTIONAL)
#we are not performing it here
```

```python
#3DATA VISUALISATION
import matplotlib.pyplot as plt
plt.scatter(df['Area'],df['Prices'])
plt.title('Area vs Prices')
plt.xlabel('Area')
plt.ylabel('Prices')
```

```python
#4 DIVIDE INTO INPUT(x) AND OUTPUT(y)
#Input(x)is always a 2 dimensional array
#Output(y)is always a 1 dimensional array
```

```python
#Input-area
#Output-prices
# : alone in row or column slicing selects all rows or columns resp
x=df.iloc[0:6,0:1].values
x
#.values convert dataframe into an array
```

```python
#in the column slicing part if there is a :,then the array is 2 dimensional
y=df.iloc[:,1].values
y
```

```python
#Train and Test variables
#We are not performing it here due to limited data
```

```python
#6 Normalisation or scaling(done only for inputs)
# Normalisation or scaling for multivariate data sets
#Here our data is univariate
```

```python
#7 Apply a classifier,REGRESSOR or clusterer
from sklearn.linear_model import LinearRegression
model=LinearRegression()#whenever a library is imported from a package,it needs to be stored into a variable
```

```python
#8 Fit the model
model.fit(x,y)
```

```python
#9 Predict the output
y_pred=model.predict(x)
y_pred
```
